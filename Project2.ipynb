# feature engineering

```python
#@title f.e 함수 { form-width: "20%" }
def engineer1(df):
#지역가중치를 위한 지역설정
  df['Bregion'] = np.nan # 블루팀의 지역이 타겟인코딩 후 높은 점수일 경우 블루팀 승리 확률 증가
  df['Rregion'] = np.nan # 레드팀의 지역이 타겟인코딩 후 높은 점수일 경우 블루팀 승리 확률 증가(Bregion의 점수와 Rregion의 점수가 다름 반대양상)
  LCK=[]
  LPL=[]
  LCS=[]
  LEC=[]
  for i in range(len(df)):
    if df['league'][i].find('LCK') != -1:
      LCK.append(df['Blue'][i])
      LCK.append(df['Red'][i])
    elif df['league'][i].find('Champions') != -1:
      LCK.append(df['Blue'][i])
      LCK.append(df['Red'][i])
    elif df['league'][i].find('LPL') != -1:
      LPL.append(df['Blue'][i])
      LPL.append(df['Red'][i])
    elif df['league'][i].find('EU_LCS') != -1:
      LEC.append(df['Blue'][i])
      LEC.append(df['Red'][i])
    elif df['league'][i].find('LEC') != -1:
      LEC.append(df['Blue'][i])
      LEC.append(df['Red'][i])
    elif df['league'][i].find('LCS') != -1:
      LCS.append(df['Blue'][i])
      LCS.append(df['Red'][i])
  LCK = list(dict.fromkeys(LCK))#LCK에 속한 팀 목록
  LPL = list(dict.fromkeys(LPL))
  LCS = list(dict.fromkeys(LCS))
  LEC = list(dict.fromkeys(LEC))
  for i in range(len(df)):
    for T in LCK:
      if df['Blue'][i] == T:
        df['Bregion'][i]='LCK'
      if df['Red'][i] == T:
        df['Rregion'][i]='LCK'
    for T in LPL:
      if df['Blue'][i] == T:
        df['Bregion'][i]='LPL'
      if df['Red'][i] == T:
        df['Rregion'][i]='LPL'
    for T in LCS:
      if df['Blue'][i] == T:
        df['Bregion'][i]='LCS'
      if df['Red'][i] == T:
        df['Rregion'][i]='LCS'
    for T in LEC:
      if df['Blue'][i] == T:
        df['Bregion'][i]='LEC'
      if df['Red'][i] == T:
        df['Rregion'][i]='LEC'
  df['Bregion'].fillna('etc',inplace=True)
  df['Rregion'].fillna('etc',inplace=True)
#날짜(string)를 날짜데이터로 변환
  for i in range(len(df)):
    a=df['date'][i]
    y=int(a[:a.find('-')])
    a=a[a.find('-')+1:]
    m=int(a[:a.find('-')])
    d=int(a[a.find('-')+1:])
    df['date'][i]=datetime.datetime(y,m,d).strftime('%Y-%m-%d')
#경기시간을 분단위로 변환
  df['time'].fillna(0,inplace=True)
  for i in range(len(df)):
    if df['time'][i] != 0:
      a=str(df['time'][i])
      m=float(a[:a.find(':')])
      s=float(a[a.find(':')+1:])/60
      m=m+s
      df['time'][i]=m
  for i in range(len(df)):
    if df['time'][i]==0:
      df['time'][i]=df['time'].mean()
#포지션별계산
  position=['T','J','M','A','S']
  for p in position:
#KDA 지표 생성
    df[f'B{p}KDA']=np.nan
    df[f'R{p}KDA']=np.nan
    for i in range(len(df)):
      if df[f'B{p}D'][i]==0:
        df[f'B{p}KDA'][i]=(df[f'B{p}K'][i]+df[f'B{p}A'])[i]*1.2 #대회공식지표에선 death가 0일 때 0으로 나누는 대신 1.2를 곱한다
      else:
        df[f'B{p}KDA'][i]=(df[f'B{p}K'][i]+df[f'B{p}A'][i])/df[f'B{p}D'][i] #kda=(kill+assist)/death
      if df[f'R{p}D'][i]==0:
        df[f'R{p}KDA'][i]=(df[f'R{p}K'][i]+df[f'R{p}A'])[i]*1.2
      else: 
        df[f'R{p}KDA'][i]=(df[f'R{p}K'][i]+df[f'R{p}A'][i])/df[f'R{p}D'][i]
    df.drop([f'B{p}K',f'B{p}D',f'B{p}A',f'R{p}K',f'R{p}D',f'R{p}A'],axis=1,inplace=True)
#골드와 CS를 분당 획득으로 변환
    df[f'B{p}G'][11913]=np.nan
    df[f'R{p}G'][11913]=np.nan
    df[f'B{p}gpm']=df[f'B{p}G']/df['time']
    df[f'B{p}cspm']=df[f'B{p}CS']/df['time']
    df[f'R{p}gpm']=df[f'R{p}G']/df['time']
    df[f'R{p}cspm']=df[f'R{p}CS']/df['time']
    df.drop([f'B{p}G',f'R{p}G',f'B{p}CS',f'R{p}CS'],axis=1,inplace=True)
#메이저리그 판수경험
    df[f'B{p}exp']=np.nan
    df[f'R{p}exp']=np.nan
    players=df[f'B{p}'].unique().tolist()+df[f'R{p}'].unique().tolist()
    players=list(dict.fromkeys(players))
    df[f'B{p}exp']=np.nan
    df[f'R{p}exp']=np.nan
    for player in players:
      df2=pd.concat([df.loc[df[f'B{p}']==player],df.loc[df[f'R{p}']==player]]).sort_values(by=['date']).reset_index()
      df2[f'exp']=df2.index
      for i in df2['index']:
        if df[f'B{p}'][i] == player:
          df[f'B{p}exp'][i]=df2['exp'].loc[df2['index']==i]
        if df[f'R{p}'][i] == player:
          df[f'R{p}exp'][i]=df2['exp'].loc[df2['index']==i]
#지금은 못써먹는 챔피언 드랍
    df.drop([f'B{p}C',f'R{p}C'],axis=1,inplace=True)
#한 리그내에서 골드지표(메타에 영향을 많이 받음)를  normalize
  league=df['league'].unique().tolist()
  df1=pd.DataFrame()
  for i in league:
    df2=df.loc[df['league']==i]
    def norm(z):
      for element in z:
        element=(element-np.mean(z))/np.std(z,ddof=1)
      return z
    for tp in ['BT','RT','BJ','RJ','BM','RM','BA','RA','BS','RS']:
      df2[f'{tp}gpm']=norm(df2[f'{tp}gpm'])
    df1=pd.concat([df1,df2])
#cs 결측치 별로 없어서 포지션별 전체평균을 넣음
  for tp in ['BT','RT','BJ','RJ','BM','RM','BA','RA','BS','RS']:
    df1[f'{tp}cspm'].fillna(np.mean(df1[f'{tp}cspm']),inplace=True)
  return df1
```

```python
#@title 데이터 랭글링 { form-width: "20%" }
#df 그대로 훈련하면 안됌 왜냐면 게임 전에 예측하려는데 게임 후 데이터를 가지고 있기 때문
#팀과 로스터로 과거에 기반한 실력지표 뽑아내기
def stats(df):
  first_day = df.set_index('league')['date'].min(level='league')
  leagues = df['league'].unique().tolist()
  record=df.drop('time',axis=1)
  df=df.drop('time',axis=1)
  for c in ['B', 'R']:
    for p in ['T','J','M','A','S']:
      for s in ['KDA','gpm', 'cspm']:
        df[f'{c}{p}{s}']=np.nan
        for i in range(len(df)):
          temp=pd.concat([record.loc[record[f'B{p}']==record[f'{c}{p}'][i]][['date',f'{c}{p}{s}']],record.loc[record[f'R{p}']==record[f'{c}{p}'][i]][['date',f'{c}{p}{s}']]])
          if len(temp[f'{c}{p}{s}'].loc[temp['date']<first_day[df['league'][i]]]) == 0:
            df[f'{c}{p}{s}'][i]=record[f'{c}{p}{s}'][i]
          else:
            df[f'{c}{p}{s}'][i]=np.mean(temp[f'{c}{p}{s}'].loc[temp['date']<first_day[df['league'][i]]])
          if len(record[f'{c}{p}exp'].loc[record['date']<first_day[df['league'][i]]]) == 0:
            df[f'{c}{p}exp'][i]=record[f'{c}{p}exp'][i]
          else:
            df[f'{c}{p}exp'][i]=record[f'{c}{p}exp'].loc[record['date']<first_day[df['league'][i]]].max()
  return df
```

```python
#CV
dists = {
    'xgbclassifier__max_depth': range(5,10),
    }
pipe = make_pipeline(
    TargetEncoder(cols=['Bregion','Rregion']),
    XGBClassifier(n_estimators=200
                  , random_state=2
                  , n_jobs=-1
                  , learning_rate=0.2
    )
)
    
clf = RandomizedSearchCV(
    pipe, 
    param_distributions=dists, 
    n_iter=50, 
    cv=3,
    scoring='f1',
    verbose=1,
    n_jobs=-1
)
clf.fit(X_train, y_train);

y_pred = clf.predict(X_test)
print('정확도: ', accuracy_score(y_test, y_pred))
print('최적 하이퍼파라미터: ', clf.best_params_)
print('f1: ', clf.best_score_)
print(classification_report(y_pred, y_test))
```
